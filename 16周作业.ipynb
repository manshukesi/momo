```python
import pandas as pd
import mytools
from scipy import stats

```

```python
认同数据表 = mytools.读取SPSS数据文件(R'data\国家认同数据清理后.sav')

cross_tab = pd.crosstab(index=[认同数据表['政治面貌'], 认同数据表['会打多少分']], 
                        columns=认同数据表['年级'], 
                        margins=True, normalize=True)  

cross_tab
```

```python
# 假设我们关注的是第二年级的学生（假设年级用数字“2”表示）
specific_grade = '大四'

# 筛选出特定年级的数据
df_specific_grade =认同数据表[认同数据表['年级'] == specific_grade]

# 创建政治面貌和会打多少分的交互表
cross_tab_specific_grade = pd.crosstab(df_specific_grade['会打多少分'], df_specific_grade['政治面貌'],margins=True,normalize=True)

# 进行卡方检验
chi2_grade, p_grade, dof_grade, expected_grade = stats.chi2_contingency(cross_tab_specific_grade)

print(chi2_grade,p_grade)
cross_tab_specific_grade
```
