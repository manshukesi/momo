```python
import pandas as pd
import mytools
from scipy import stats

```

```python
认同数据表 = mytools.读取SPSS数据文件(R'data\国家认同数据清理后.sav')

cross_tab = pd.crosstab(index=[认同数据表['政治面貌'], 认同数据表['世博会']], 
                        columns=认同数据表['年级'], 
                        margins=True, normalize=True)  

cross_tab
```

```python
# 假设我们关注的是第二年级的学生（假设年级用数字“2”表示）
specific_grade = '大二'

# 筛选出特定年级的数据
df_specific_grade =认同数据表[认同数据表['年级'] == specific_grade]

# 创建政治面貌和会打多少分的交互表
cross_tab_specific_grade = pd.crosstab(df_specific_grade['政治面貌'], df_specific_grade['世博会'],margins=True,normalize=True)

# 进行卡方检验
chi2_grade, p_grade, dof_grade, expected_grade = stats.chi2_contingency(cross_tab_specific_grade)

print(chi2_grade,p_grade)
cross_tab_specific_grade
```

```python
mytools.两个无序类别变量的统计分析(df_specific_grade,'政治面貌','世博会')
```

分析：
从数据中可以看出，团员占比最高，其次是党员，然后是群众，最后是其他。
团员：观测值显示出团员对世博会的态度普遍较好，尤其是在“比较好”和“感到自豪”这两个积极类别上。期望值在这些类别上也较高，但观测值显著高于期望值，表明团员群体对世博会态度更为积极。
党员：党员的观测值在“没感觉”和“一般”这两个较为中立的态度上高于期望值，在“比较好”和“感到自豪”上低于期望值。表明党员群体对世博会态度相对中立。
群众：群众的观测值在“没感觉”和“一般”上与期望值接近，在“比较好”上略高于期望值，在“感到自豪”上显著高于期望值。这表明群众中有一部分人对世博会持有较为积极的态度。
其他：这个群体的人数较少，从数据上看，他们对世博会的态度没有明显倾向。
All：可以看出大多数人选择了“比较好”或“感到自豪”这两个较为积极的态度。
