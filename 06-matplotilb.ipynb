```python
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"]=["SimHei"] #设置字体
plt.rcParams["figure.figsize"]=(6.4,4.8) #设置大小，默认为英寸
plt.rcParams["axes.unicode_minus"]=False #该语句解决图像中的“-”负号乱码问题
```

```python
import matplotlib.pyplot as plt  
import numpy as np  
  
# 创建数据  
x = np.linspace(0, 10, 100)  
y = np.sin(x)  
  
# 创建一个新的图形，并设置大小  
plt.figure(figsize=(10,6))  
  
# 在图形中添加一条线  
plt.plot(x, y, label='sin(x)')  
  
# 设置图形的标题以及x和y轴的标签  
plt.title('Sine Wave')  
plt.xlabel('x (0 to 10)')  
plt.ylabel('sin(x)')  
  
# 添加图例  
plt.legend()  
  
# 显示图形  
plt.show()
```

```python
import matplotlib.pyplot as plt  
import numpy as np  
x = np.linspace(-2, 2, 1000)  
y1 = np.sqrt(1-(abs(x)-1)**2)  
y2 = -3*np.sqrt(1-(abs(x)/2)**0.5)  
  
fig, ax = plt.subplots()  
ax.fill_between(x, y1, color='red')  
ax.fill_between(x, y2, color='red')  
ax.set_aspect('equal')  
  
plt.show()
```

```python
import matplotlib.pyplot as plt  
import numpy as np  
  
# 数据  
x = np.random.rand(50) * 10  
y = np.random.rand(50) * 10  
  
# 创建散点图  
plt.scatter(x, y)  
  
# 添加标题和轴标签  
plt.title('Scatter Plot')  
plt.xlabel('x')  
plt.ylabel('y')  
  
# 显示图像  
plt.show()
```

```python
import matplotlib.pyplot as plt  
import numpy as np  
  
# 数据  
x = np.arange(5)  
y = np.random.rand(5)  
  
# 创建柱状图  
plt.bar(x, y)  
  
# 添加标题和轴标签  
plt.title('Bar Plot')  
plt.xlabel('Category')  
plt.ylabel('Value')  
  
# 显示图像  
plt.show()
```

```python
import matplotlib.pyplot as plt  
import numpy as np  
  
# 数据  
labels = ['A', 'B', 'C', 'D']  
sizes = [15, 30, 45, 10]  
  
# 创建饼图  
plt.pie(sizes, labels=labels)  
  
# 添加标题和轴标签（虽然饼图没有明显的轴）  
plt.title('Pie Chart')  
plt.ylabel('Population')  #尽管没有y轴，但仍然可以添加标签以保持一致性。  
plt.show()  # 显示图像，也可以使用 plt.savefig() 来保存图像。
```
